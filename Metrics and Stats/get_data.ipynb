```python
import datetime

import matplotlib.pyplot as plt
import numpy as np
import yfinance as yfin
```

```python
start = datetime.date.today() - datetime.timedelta(days=5*365)
end = datetime.date.today()
df = yfin.download(["META", "F", "BTC-USD"], start, end)["Adj Close"]
# df["AMZN"] = 20 * df["AMZN"]  # code adapted for recent AMZN stock split

df.head(10)
```

```python
df.describe()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()
ax3 = ax1.twinx()

df["META"].plot(ax=ax1, y="META" ,color="r", label="META", legend=True)
df["F"].plot(ax=ax2, y="F",color="g", label="F", legend=True)
df["BTC-USD"].plot(ax=ax3,y="BTC-USD", color="b", label="BTC-USD", legend=True)

ax1.set_ylabel("META", color="r")
ax2.set_ylabel("F", color="g")
ax3.set_ylabel("BTC-USD", color="b")
ax3.spines["right"].set_position(("outward", 60))




# Separate legend positions
ax1.legend(["META"],loc='lower right', bbox_to_anchor=(0.5, 0.5))
ax2.legend(["F"],loc='lower right', bbox_to_anchor=(0.5, 0.4))
ax3.legend(["BTC-USD"],loc='lower right', bbox_to_anchor=(0.5, 0.3))

plt.show()
```
